```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline
```

```python
df=pd.read_csv("G:\sick_csv.csv")
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#dropping unwanted columns
df.drop(['age','sex','TBG','referral_source','Class'],axis='columns',inplace=True)
df.head(15)
```

```python
#converting cat to num
df['on_thyroxine'] = df['on_thyroxine'].map({'t':1 ,'f':0}) 
df['query_on_thyroxine'] = df['query_on_thyroxine'].map({'t':1 ,'f':0}) 
df['on_antithyroid_medication'] = df['on_antithyroid_medication'].map({'t':1 ,'f':0}) 
df['sick'] = df['sick'].map({'t':1 ,'f':0}) 
df['pregnant'] = df['pregnant'].map({'t':1 ,'f':0}) 
df['I131_treatment'] = df['I131_treatment'].map({'t':1 ,'f':0}) 
df['lithium'] = df['lithium'].map({'t':1 ,'f':0}) 
df['goitre'] = df['goitre'].map({'t':1 ,'f':0}) 
df['tumor'] = df['tumor'].map({'t':1 ,'f':0}) 
df['hypopituitary'] = df['hypopituitary'].map({'t':1 ,'f':0}) 
df['psych'] = df['psych'].map({'t':1 ,'f':0}) 
```

```python
df.head()
```

```python
#dropping measured columns as they doesn't account for whether a person sick or not
df.drop(['TSH_measured','T3_measured','TT4_measured','T4U_measured','FTI_measured','TBG_measured'],axis='columns',inplace=True)
df.head()
```

```python
df['thyroid_surgery'] = df['thyroid_surgery'].map({'t':1 ,'f':0}) 
df['query_hypothyroid'] = df['query_hypothyroid'].map({'t':1 ,'f':0}) 
df['query_hyperthyroid'] = df['query_hyperthyroid'].map({'t':1 ,'f':0}) 
df.head()
```

```python
# now replacing Nan values with mean of respective column
df['TSH'].fillna((df['TSH'].mean()), inplace=True)
df['T3'].fillna((df['T3'].mean()), inplace=True)
df['TT4'].fillna((df['TT4'].mean()), inplace=True)
df['T4U'].fillna((df['T4U'].mean()), inplace=True)
df['FTI'].fillna((df['FTI'].mean()), inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
#here in order to continue with classification we must choose independent variables and dependent varaiables here i choose
#sick as my target variable and rest as features
df1=df['sick']
df1.head()
```

```python
df.drop('sick',axis='columns',inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X=df
X.head(2)
```

```python
y=df1
y.head(2)
```

```python
#now we divide whole dataset into train and test set to train and evaluae model
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3)
```

```python
cls=LogisticRegression()
cls.fit(X_train,y_train)
```

```python
cls.predict(X_test)
```

```python
cls.score(X_test,y_test)
```
